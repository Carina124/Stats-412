```python
df = pd.read_csv("food_dataset.csv") #fortunately there's no text data
df
```

```python
#making the categorical variables consistent
df.loc[df['edu'] == 1, 'edu'] = 0
df.loc[df['edu'] == 2, 'edu'] = 1
df.loc[df['edu'] == 3, 'edu'] = 2

df.loc[df['income'] == 1, 'income'] = 0
df.loc[df['income'] == 2, 'income'] = 1
df.loc[df['income'] == 3, 'income'] = 2

df.loc[df['race'] == 1, 'race'] = 0
df.loc[df['race'] == 2, 'race'] = 1
df.loc[df['race'] == 3, 'race'] = 2
df.loc[df['race'] == 4, 'race'] = 3

df.loc[df['hh_size'] == 1, 'hh_size'] = 0
df.loc[df['hh_size'] == 2, 'hh_size'] = 1
df.loc[df['hh_size'] == 3, 'hh_size'] = 2
df.loc[df['hh_size'] == 4, 'hh_size'] = 3

df.loc[df['marital_status'] == 1, 'marital_status'] = 0
df.loc[df['marital_status'] == 2, 'marital_status'] = 1
df.loc[df['marital_status'] == 3, 'marital_status'] = 2
df.loc[df['marital_status'] == 4, 'marital_status'] = 3

df
```

```python
#subset children, marital_status, employed_new, urban, hh_size, edu, race, income
subdf = df.iloc[:, 13:21]
subdf = subdf.drop('age', axis=1)
subdf
```

```python
#counts and percentages of categorical variables
edu_count = subdf.groupby("edu")["edu"].count()
income_count = subdf.groupby("income")["income"].count()
race_count = subdf.groupby("race")["race"].count()
hh_size_count = subdf.groupby("hh_size")["hh_size"].count()
marital_status_count = subdf.groupby("marital_status")["marital_status"].count()
children_count = subdf.groupby("children")["children"].count()
employed_new_count = subdf.groupby("employed_new")["employed_new"].count()


df2 = pd.DataFrame({
    "edu": edu_count,
    "income": income_count,
    "race": race_count,
    "hh_size" : hh_size_count,
    "marital_status": marital_status_count,
    "children": children_count,
    "employed_new": employed_new_count
})

df2["edu_percent"] = None
df2["income_percent"] = None
df2["race_percent"] = None
df2["hh_size_percent"] = None
df2["marital_status_percent"] = None
df2["children_percent"] = None
df2["employed_new_percent"] = None

df2["edu_percent"] = (df2["edu"] / 22448) * 100
df2["income_percent"] = (df2["income"] / 22448) * 100
df2["race_percent"] = (df2["race"] / 22448) * 100
df2["hh_size_percent"] = (df2["hh_size"] / 22448) * 100
df2["marital_status_percent"] = (df2["marital_status"] / 22448) * 100
df2["children_percent"] = (df2["children"] / 22448) * 100 
df2["employed_new_percent"] = (df2["employed_new"] / 22448) * 100 


df2 = df2[['edu', 'edu_percent', 'income', 'income_percent', 'race', 'race_percent', 'hh_size', 'hh_size_percent', 'marital_status', 'marital_status_percent', 'children', 'children_percent', "employed_new", "employed_new_percent"]]

df2

```

```python
#creating column that will sort the zip codes into regions
df["zip_region"] = None

mask = df['zip'].astype(str).str.match(r'^[98]\d{4}$') #west
mask2 = df['zip'].astype(str).str.match(r'^[456]\d{4}$') #midwest
mask3 = df['zip'].astype(str).str.match(r'^[37]\d{4}$') #south
mask4 = df['zip'].astype(str).str.match(r'^[01]\d{4}$') #northeast
mask5 = df['zip'].astype(str).str.match(r'^[1]\d{3}$') #northeast

df.loc[mask, 'zip_region'] = 'west'
df.loc[mask2, 'zip_region'] = 'midwest'
df.loc[mask3, 'zip_region'] = 'south'
df.loc[mask4, 'zip_region'] = 'northeast'
df.loc[mask5, 'zip_region'] = 'northeast'
df
```
